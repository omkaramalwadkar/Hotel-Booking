<b><i>Problem Statement</i></b>
</div>
 In the recent years, City Hotels and Resort Hotels have seen very high cancellation rates. Each hotel is now dealing with large number of issues such as less revenues, unused hotel rooms due to it’s cancellation etc. So lowering the cancellation rates of the hotel rooms is the primary goal for both the hotels to increase their efficiency in generating revenues and even prevent the hotel rooms from being unutilized. <div style="text-align: center; font-size: 15px;">
<b><i>About Data</i></b>
</div>

Columns:

1. **hotel**: One of the hotels is a resort hotel and the other is a city hotel.

2. **is_canceled**: Value indicating if the booking was canceled (1) or not (0).

3. **lead_time**: Number of days between booking date and arrival date.

4. **arrival_date_year**: Year of arrival date.

5. **arrival_date_month**: Month of arrival date with 12 categories: “January” to “December”.

6. **arrival_date_week_number**: Week number of the arrival date.

7. **arrival_date_day_of_month**: Day of the month of the arrival date.

8. **stays_in_weekend_nights**: Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel.

9. **stays_in_week_nights**: Number of week nights (Monday to Friday) the guest stayed.

10. **adults**: Number of adults.

11. **children**: Number of children.

12. **babies**: Number of babies.

13. **meal**: Type of meal booked. (e.g., BB – Bed & Breakfast)

14. **country**: Country of origin.

15. **market_segment**: Market segment designation. (e.g., "TA" means "Travel Agents" and "TO" means "Tour Operators")

16. **distribution_channel**: Booking distribution channel. (e.g., "TA" means "Travel Agents" and "TO" means "Tour Operators")

17. **is_repeated_guest**: Value indicating if the booking name was from a repeated guest (1) or not (0).

18. **previous_cancellations**: Number of previous bookings that were cancelled by the customer prior to the current booking.

19. **previous_bookings_not_canceled**: Number of previous bookings not cancelled by the customer prior to the current booking.

20. **reserved_room_type**: Code of room type reserved.

21. **assigned_room_type**: Code for the type of room assigned to the booking.

22. **booking_changes**: Number of changes/amendments made to the booking.

23. **deposit_type**: Type of deposit made for the booking.

24. **agent**: ID of the travel agency that made the booking.

25. **company**: ID of the company/entity that made the booking or responsible for paying the booking.

26. **days_in_waiting_list**: Number of days the booking was in the waiting list before it was confirmed to the customer.

27. **customer_type**: Type of customer. (e.g., Group, Transient, Transient-party)

28. **adr**: Average Daily Rate.

29. **required_car_parking_spaces**: Number of car parking spaces required by the customer.

30. **total_of_special_requests**: Number of special requests made by the customer.

31. **reservation_status**: Current reservation status. (e.g., Check-Out, No-Show)

32. **reservation_status_date**: Date at which the last status was set.

33. **name**: Name of the Guest (Not Real)

34. **email**: Email (Not Real)

35. **phone-number**: Phone number (Not Real)

36. **credit_card**: Credit Card Number (Not real)

<b><i>Objective</i></b>
</div>
The main objective of the hotel booking Cancellation project is to find out the factors for which so many hotel booking cancellations are being taken place. Based on that, we need to provide many good solutions to further prevent the large number of cancellation of the hotel rooms.

 <div style="text-align: center; font-size: 15px;">
<b><i>Assumptions</i></b>
</div>

(i) No extraordinary events occurred between 2015 and 2017 that would significantly impact the dataset used for analysis.

(ii) The data provided remains relevant and can be effectively utilized to inform the hotel's strategic planning.

(iii) Assumption is made that there will be no adverse reactions or pushback from hotels regarding the solutions proposed to enhance their revenues.

(iv) It's assumed that the hotels are not currently implementing any of the suggested solutions.

(v) The primary driver impacting hotel revenues is identified as the cancellation of booked hotel rooms.

(vi) Cancellations result in hotel rooms remaining vacant for the duration of the originally booked stay.

(vii) Clients have the option to cancel their hotel reservations within the same calendar year in which the reservations were made.

<div style="text-align: center; font-size: 15px;">
<b><i>Research Questions</i></b>
</div>

(i) What factors contribute to hotel reservation cancellations?

(ii) What strategies can be implemented to improve hotel reservation cancellation management?

(iii) How can hotels be supported in making informed pricing and promotional decisions?


<div style="text-align: center; font-size: 15px;">
<b><i>Hypothesis</i></b>
</div>

(i) Cancellation of hotel room bookings occurs more frequently when their prices are high.

(ii) Customers tend to cancel more frequently when there is a longer waiting list.

(iii) The majority of clients make their hotel room reservations through offline travel agents.

```python
df = df[df['adr'] < 5000]
```

```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)
plt.figure(figsize = (5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Cancelled','Cancelled'],df['is_canceled'].value_counts(),edgecolor = 'k', width=0.7)
plt.show()
```

<div style="text-align: center; font-size: 15px;">
<b><i>Observation</i></b>
</div>

The bar graph above displays the percentage of hotel bookings that were canceled compared to those that were not. It's evident that 62% of the booked hotel rooms remain unaffected by cancellations. However, a notable 37% of the reserved hotel rooms have been canceled, significantly impacting the hotel's earnings

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue='is_canceled', data = df, palette='Blues')
legend_lables,_=ax1 .get_legend_handles_labels()
ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status In Different Hotels',size=20)
plt.xlabel('Hotel')
plt.ylabel('Number Of Reservations')
plt.legend(['Not Cancelled' , 'Cancelled'])
plt.show()
```

<div style="text-align: center; font-size: 15px;">
<b><i>Observation</i></b>
</div>

(i) The bar graph above compares the ratio of cancellations to non-cancellations for both the Resort and City hotels.

(ii) We observe that the ratio of non-cancellations to cancellations is significantly higher for the Resort Hotel compared to the City Hotel.

(iii) Additionally, the cancellation rate is higher for the City Hotel compared to the Resort Hotel, possibly due to the lower pricing of the City Hotel in comparison to the more expensive Resort Hotel.

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts( normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts( normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(22,9))
plt.title('Average Daily Rate In Resort And City Hotel', fontsize=25)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

<div style="text-align: center; font-size: 15px;">
<b><i>Observation</i></b>
</div>


(i) The graph shows consistent data trends from January 2016 to January 2017.

(ii) On weekends, there are noticeable spikes, suggesting that the average daily rate (price) increases significantly on Saturdays and Sundays.

(iii) The average daily rate (ADR) for the Resort Hotel is notably higher than that of the City Hotel, indicating that the Resort Hotel is more expensive than the City Hotel.Hotel.

```python
df['month']= df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x = 'month' , hue='is_canceled' , data=df , palette= 'bright')
legend_labels,_= ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Resrvation Status Per Month',size=20)
plt.xlabel('month')
plt.ylabel('number of reservation')
plt.legend(['not_canceled','canceled'])
plt.show()
```

<div style="text-align: center; font-size: 15px;">
<b><i>Observation</i></b>
</div>


(i) The bar graph displays the reservation status of City and Resort Hotels each month.

(ii) January experienced the highest number of hotel cancellations and the lowest number of reservations.

(iii) In contrast, August saw the fewest hotel cancellations and the highest number of reservations.

```python
cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_data = df[df['is_canceled']==0]
not_cancelled_df_adr=not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label='cancelled')
plt.legend(fontsize=20)
plt.show()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr[ 'reservation_status_date']> '2016') & (cancelled_df_adr['reservation_status_date']> '2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr[ 'reservation_status_date']> '2016') & (not_cancelled_df_adr[ 'reservation_status_date']<'2017-09')]
```

<div style="text-align: center; font-size: 15px;">
<b><i>Observation</i></b>
</div>


(i) The graph above compares the Average Daily Rate (ADR) values for canceled and non-canceled hotel rooms.

(ii) It's evident from the graph that when the ADR is high, there are more cancellations compared to instances where the booking is not canceled.

(iii) This analysis suggests that higher prices are associated with a greater likelihood of hotel room cancellations.
